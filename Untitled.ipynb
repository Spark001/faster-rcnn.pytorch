```python
import os
import sys
import scipy.io as sio
sys.path.insert(0, '/home/disk/workspace')
from Library.pascal_voc_io import PascalVocWriter
import cv2
import pdb
import h5py
from tqdm import tqdm
```

```python
ImgDIR = '/home/disk/dataset/voc/VOC/VOCdevkit2007/VOC2007/JPEGImages'
EBfile = '/home/disk/dataset/EdgeBoxes70.h5'
edgebox = h5py.File(EBfile, 'r')
# print (type(edgebox))
imgnames = edgebox.keys()
norm_box = True
edgeboxdict = dict()
tol = 0
for imgname in tqdm(imgnames):
    if tol<5000:
        tol +=1 
        continue
    imgpath = os.path.join(ImgDIR, imgname+'.jpg')
    if not os.path.exists(imgpath):
        print(imgpath,'not exist!')
        continue
    img = cv2.imread(imgpath)
    assert img.shape, imgpath
    height,width,_ = img.shape
    boxes = []
    for _,i in enumerate(edgebox[imgname]):
#         print (type(edgebox[imgname]))
#         if float(i[-1]) < 0.1: continue
        x1,y1,x2,y2,score = float(i[0]),float(i[1]),float(i[2]),float(i[3]),float(i[4])
        if norm_box:
            x1 *= width
            x2 *= width
            y1 *= height
            y2 *= height
            x1,y1,x2,y2 = map(round, [x1,y1,x2,y2])
        cls = 'objectness'
        diff = 0
        boxes.append([x1,y1,x2,y2,cls,score])
    edgeboxdict[imgname] = boxes
```
